```python
import pandas
import pandas as pd
```

```python
bookcrossing = pd.read_csv('../dataset/BookCrossing/Preprocessed_data.csv')
bookcrossing.rename(columns= {'book_title':'Title'}, inplace=True)
bookcrossing.head(2)
```

```python
amazonbook = pd.read_csv('../dataset/AmazonBook/books_data.csv')
#amazonbook.head(2)
```

```python
review = pd.read_csv('../dataset/AmazonBook/Books_rating.csv')
#review.head(2)
```

```python
review.drop(columns=['Id', 'Price', 'profileName', 'review/time', 'review/helpfulness'], inplace=True)
#review.head(2)
```

```python
data = pd.merge(amazonbook, bookcrossing, on='Title', how='inner')
#data.head(2)
```

```python
data.dropna(inplace=True)
```

```python
data = data[data['Language'] != '9']
#data.head(2)
```

```python
data.drop(columns=['image', 'previewLink', 'infoLink', 'categories', 'Unnamed: 0', 'isbn', 'age', 'img_s', 'img_m', 'img_l', 'city', 'state', 'country', 'location', 'year_of_publication', 'user_id', 'rating'], inplace=True)
```

```python
data['publishedDate'] = data['publishedDate'].str.extract(r'(\d{4})')
```

```python
#data.head(2)
```

```python
data['Publishers'] = data['publisher_x'].fillna(data['publisher_y'])
```

```python
data.drop(['publisher_x', 'publisher_y'], axis=1, inplace=True)
```

```python
len(data['Title'].unique())
```

```python
data = data.drop_duplicates()
#data.head(2)
```

```python
#review.head(2)
```

```python
data_withReviews = pd.merge(data, review, on='Title', how='inner')
#data_withReviews.head(2)
```

```python
data_withReviews.dropna(inplace=True)
```

```python
#data_withReviews.head(2)
```

```python
data_withReviews.reset_index()
data_withReviews.to_csv('reviews1.csv', index=False)
```
